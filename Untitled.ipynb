```python
import sounddevice as sd
import numpy as np
import scipy.io.wavfile as wav
import time
from bokeh.io import push_notebook, show, output_notebook
from bokeh.layouts import row
from bokeh.plotting import figure
from bokeh.palettes import BuPu, Greens, Set3
from scipy.fftpack import fft, rfft
import math
```

```python
output_notebook()
opts = dict(plot_width=1000, plot_height=400, min_border=0)
```

```python
p2 = figure(**opts)

fileName = 'combined_all_minus_orig.data'

f = open(fileName,'r')
data = np.loadtxt(f, delimiter=' ')
f.close()

xLen = range(len(data[0]))[:2048]


for i in range(len(data)):
    
    dataSep = data[i].tolist()[:2048]

    if i < 15:
        p2.line(xLen, dataSep, line_width=1, color='blue', alpha=0.1)
    else:
        
        for k in range(len(dataSep)):
            dataSep[k] = dataSep[k] + 1
        
        p2.line(xLen, dataSep, line_width=1, color='green', alpha=0.1)

show(p2)
```

```python
plot = figure()
plot.patches(xs = [[1,1,2,2],
                   [2,2,3,3]],ys = [[1,2,1,2],
                                    [1,2,1,2]],fill_color = ["purple","red"],line_color = ["black","black"])
show(plot)

[1,1,2,2]
[1,2,1,2]
```
